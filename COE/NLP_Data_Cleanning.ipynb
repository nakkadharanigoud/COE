NLP Technique

Step 1: Lowercasing

```python
text="I Love NLP"
lower_text=text.lower()
print(lower_text)

title_text=text.title()
print(title_text)

upper_text=text.upper()
print(upper_text)
```

###step 2: Remove Punctuation
-remove symbols like . , ! ; :

```python
import string
text="Wow!!! This is amazing. NLP is great,isn't it?"

#Remove punctuation
clean_text=text.translate(str.maketrans('','',string.punctuation))
print(clean_text)
```

Step 3:Tokenization
-split text into smaller units(words or sentences)

```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt_tab', quiet=True)
text="Wow!!! This is amazing. NLP is great,isn't it?"
tokens=word_tokenize(text)
print(tokens)
```

Step4: Remove the Stopwords
-remove common words that do not add the strong meaning

```python
from nltk.corpus import stopwords
nltk.download('stopwords', quiet=True)
text="Wow!!! This is amazing. NLP is great,isn't it?"
tokens=word_tokenize(text)

stop_words=set(stopwords.words('english'))
```

```python
#stemming
from nltk.stem import PorterStemmer
text="NLP is amazing. I love learning NLP."
tokens=word_tokenize(text)

stemmer=PorterStemmer()
stemmed_tokens=[stemmer.stem(token) for token in tokens]
print(stemmed_tokens)
```

###Step 6: Lemmatization
- covert word to its dictionary base form

```python
#lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet', quiet=True)
text="NLP is amazing. I love learning NLP."
lemmatizer=WordNetLemmatizer()
lemmatized_tokens=[lemmatizer.lemmatize(word) for word in tokens]
print(lemmatized_tokens)
```

###Step 7: Remove Number
-Remove numeric value if not required

```python
import re
text="Price: $100.00"
 #Remove leading/trailing whitespace
clean_text=text.strip()
print(clean_text)
#Remove number
clean_text=re.sub(r'\d+', '', clean_text)
print(clean_text)
```

Remove URL
-remove web links form text

```python
import re
#link text
text="Visit our website at https://www.example.com for more info."
clean_text=re.sub(r'http\S+', '', text)
print(clean_text)
```

###Step 9: Remove HTML TAG
-remove html markup

```python
#import lib html tag remove
import re
text="<p>This is a paragraph.</p>"
clean_text=re.sub(r'<.*?>', '', text)
print(clean_text)
```

###Step 11: Bag of word
-count the frequency of each words in vocabulary

```python
#import lib bag of word
from sklearn.feature_extraction.text import CountVectorizer
sentences=["I love NLP","NLP is amazing","I enjoy learning NLP"]
vectorizer=CountVectorizer()
X=vectorizer.fit_transform(sentences)
print("vocabulary:" ,vectorizer.get_feature_names_out())
print("Bag of Words Matrix:\n", X.toarray())
```

###Step 12

```python
#import TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
sentences=["I love NLP","NLP is amazing","I enjoy learning NLP"]
vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(sentences)
print("vocabulary:" ,vectorizer.get_feature_names_out())
print("TF-IDF Matrix:\n", X.toarray())
```
