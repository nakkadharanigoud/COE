```python
#import required lib
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN,Dense,Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
# sample dataset
data={
    'text':[
        "I love this movie",
        "This film is terrible",
        "amazing acting and story",
        "I hate this movie",
        "best movie ever",
        "worst movie ever",
        "really good film",
        "not good film"
    ],
    'sentiment':[1,0,1,0,1,0,1,0]
}
```

```python
# convert dict to dataframe
df=pd.DataFrame(data)
df
```

```python
#tokenization
Tokenizer=Tokenizer(num_words=1000)
Tokenizer.fit_on_texts(df['text'])
```

```python
# convert text to sequences
sequences = Tokenizer.texts_to_sequences(df['text'])
```

```python
#padding
X=pad_sequences(sequences,maxlen=5)
y=df['sentiment']
```

```python
#split data into train and test
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
##RNN Model
model=Sequential()
model.add(Embedding(input_dim=1000,output_dim=32,input_length=5))
model.add(SimpleRNN(32))
model.add(Dense(units=1,activation='sigmoid'))
```

```python
#compile the model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#train model Simple RNN
model.fit(X_train,y_train,epochs=10,validation_data=(X_test,y_test),batch_size=2)
```

```python
# Evaluate the model
loss, acc = model.evaluate(X_test, y_test)
print("Test Loss:", loss)
print("Test Accuracy:", acc)
```
