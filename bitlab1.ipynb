```python
a={'Mumbai': 50, 'Hyd': 40, 'delhi': 30, 'Dubai': 20, 'Jaipur': 100, 'Kolkata': 350, 'chennai': 280, 'Dadar': 200}
def city(a):
  l=[]
  for i in a:
    if i in a:
      if a[i]<100:
        l.append(i)
  return l
print(city(a))
```

```python
def vowel(a):
  b=['a','e','i','o','u','A','E','I','O','U']
  c=0
  for char in a:
    if char not in b:
      c=c+1
  return c
a=input()
print(vowel(a))
```

```python
def count(a):
  v=0
  c=0
  b=['a','e','i','o','u','A','E','I','O','U']
  for char in a:
    if char in b:
      v=v+1
    else:
      c=c+1
  return {'vowel':v,'consonant':c}
a=input()
print(count(a))
```
