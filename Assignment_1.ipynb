```python
#rename all columns in uppercase
df=df.rename(columns={'START_DATE*':'START_DATE','END_DATE*':'END_DATE','START*':'START','CATEGORY*':'CATEGORY','STOP*':'STOP','MILES*':'MILES','PURPOSE*':'PURPOSE'})
df
```

```python
#display rides where CATEGORY=='Business'
business=df[df['CATEGORY']=='Business']
business.head()
```

```python
#5 rides with the longest distance(MILES)
df.nlargest(5,'MILES')
```

```python
#replace all missing PURPOSE values with 'NOT SPECIFIED'
df['PURPOSE']=df['PURPOSE'].fillna('NOT SPECIFIED')
df.head()
```

```python
#create a new column TRIP_DURATION using END_DATE-START_DATE
# Remove the row containing 'Totals' in the 'START_DATE' column
df = df[df['START_DATE'] != 'Totals'].copy()
df['START_DATE']=pd.to_datetime(df['START_DATE'])
df['END_DATE']=pd.to_datetime(df['END_DATE'])
df['TRIP_DURATION']=df['END_DATE']-df['START_DATE']
df.head()
```

```python
#sort trips by distance in descending order
df.sort_values(by='MILES',ascending=False)
```
